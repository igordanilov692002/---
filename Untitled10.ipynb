```python
missing_values_count = data_rfc.isnull().sum()
missing_values_count
```

# Анализ N(процент пропуска)

рассмотрим разное значение для пропуска в данных

```python
import matplotlib.pyplot as plt

percent_values = [30, 50, 70, 90]
fig, axs = plt.subplots(len(percent_values), figsize=(10, 10))
missing_values_percent = (missing_values_count / len(data_rfc)) * 100
for i, percent_value in enumerate(percent_values):
    filtered_columns = missing_values_percent[missing_values_percent < percent_value]

    filtered_columns.plot(kind='bar', ax=axs[i])
    axs[i].set_title(f'Теги с пропусками < {percent_value}%')

plt.subplots_adjust(hspace=20)
plt.tight_layout()
plt.show()
```

# ***Скрипт*** субъект - признаки

цель данного скрипта
получение json файла, в котором ключом является amunity(субъект), значения - теги

```python
def obj2tags(df, percent_value, tag='amenity'):
  key2tags = {}
  for i in df[tag].unique():
    choose_cat = df[df[tag] == i]
    missing_values_count = choose_cat.isnull().sum()
    missing_values_percent = (missing_values_count / len(choose_cat)) * 100
    filtered_columns = missing_values_percent[missing_values_percent < percent_value]
    key2tags[i] = filtered_columns.index.tolist()
  return key2tags
```

```python
dict_data = obj2tags(data, percent_value=50, tag='amenity')
dict_data['fast_food']
```

```python
file_path = 'obj2tags.json'
with open(file_path, 'w') as f:
  json.dump(dict_data, f, indent=4)  #  indent=4 - количество пробелов
  print(f"Словарь успешно сохранен в файл {file_path}.")
```

# Вывод

При анализе признаков в зависимости от повышения значения процента пропуска можно заметить следующие тенденции:

При низких значениях процента пропуска (например, менее 10%), наиболее часто встречаются признаки, которые часто заполняются или несут основную информацию о объекте (координаты, наименование, часы для посещения).

По мере увеличения значения процента пропуска (например, более 30%), начинают появляться признаки, которые могут быть менее важными или менее заполненными для большинства объектов (например, информация о бренде, ссылки на веб-сайты, доступ в интернет).

При очень высоких значениях процента пропуска (например, более 90%), встречаются признаки, которые, вероятно, редко заполняются или не являются обязательными для большинства объектов. Возможно, это связано с тем, что эти признаки представляют собой более специфическую информацию о конкретных типах объектов или предпочтениях пользователей (информация о платежных системах, уровень доступа к объекту, наличие внешнего сидения).

Таким образом, анализ признаков в зависимости от процента пропуска может помочь понять, какие признаки имеют наибольшее значение или стабильно заполняются.
